```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
file_path=r"C:\Users\Venks\Desktop\Project\Langchain_Ops\PDF_to_JSON\data\raw_pdf\[Fund Factsheet - May]360ONE-MF-May 2025.pdf"
```

```python
pdf_loader=PyPDFLoader(file_path)
pdf_docs=pdf_loader.load()
```

```python
pdf_docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

pdf_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=pdf_splitter.split_documents(pdf_docs)
final_documents
```

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
ollama_embeddings = OllamaEmbeddings(model="all-minilm:latest")# By default it uses llama2 else OllamaEmbeddings("gemma:2b")
```

```python
ollama_embeddings
```

```python
embeded_pdf_doc=ollama_embeddings.embed_documents(final_documents)
```

```python
from langchain_chroma import Chroma
```

```python
db = Chroma(
    collection_name="Test_PDF_collection",
    embedding_function=embeded_pdf_doc,
    persist_directory="./chroma_langchain_db",
)
```

```python
db.add_documents(documents=embeded_pdf_doc,id=page)
```

```python
query="What is the all-time high dividend transfer which was made to the Government of India?"
retrieved_results= db.similarity_search(query)
```
